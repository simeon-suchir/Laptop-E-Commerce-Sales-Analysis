```python
import pandas as pd
df  = pd.read_csv("laptops_final.csv")
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Brand')
plt.xticks(rotation=45)
plt.title('Count of Laptops by Brand')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Processor-Brand', order=df['Processor-Brand'].value_counts().index)
plt.title('Count of Processor Brands')
plt.xlabel('Processor Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Brand', hue='Processor-Brand')
plt.title('Processor Brands by Laptop Brand')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Price', y='Rating', alpha=0.7)
plt.title('Price vs Rating Correlation')
plt.xlabel('Price')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

```python
colormap = cm.get_cmap("tab20b") 
colors = [colormap(i / 6) for i in range(6)]
plt.figure(figsize=(15, 12))
processor_dist = df['Processor-Brand'].value_counts()
plt.pie(processor_dist.values, labels=processor_dist.index,textprops={'fontsize': 7}, autopct='%1.1f%%',colors=colors)
plt.title('Distribution of Processor Brands')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
avg_prices = df.groupby('Processor-Brand')['Price'].mean().sort_values(ascending=False)
plt.bar(avg_prices.index, avg_prices.values)
plt.title('Average Laptop Price by Processor Brand')
plt.xticks(rotation=45)
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Brand', y='Price', hue='Processor-Brand')
plt.title('Price Distribution by Brand and Processor')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='RAM', hue='Brand')
plt.title('Brand Distribution by RAM')
plt.xticks(rotation=45)
plt.show()
```

```python
threshold = 0.0055

colormap = cm.get_cmap("tab20b") 
colors = [colormap(i / 11) for i in range(11)]

brand_counts = df['Brand'].value_counts()
brand_percentages = brand_counts / brand_counts.sum()
cleaned_brands = brand_percentages[brand_percentages >= threshold]
others_percentage = brand_percentages[brand_percentages < threshold].sum()
cleaned_brands['Others'] = others_percentage
plt.figure(figsize=(10, 10))
plt.pie(cleaned_brands.values, labels=cleaned_brands.index, autopct='%1.1f%%', startangle=140,textprops={'fontsize': 7},colors=colors)
plt.title('Distribution of Laptop Brands')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='Hard-Disk', y='Price')
plt.xticks(rotation=45)
plt.title('Distribution of Laptop Prices by Hard Disk Size')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.kdeplot(data=df, x='Price', hue='Brand', fill=True, common_norm=False, alpha=0.5)
plt.title('Density Plot of Laptop Prices Grouped by Brand')
plt.xlabel('Price')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Price'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Laptop Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
correlation_matrix = df[['Price', 'Rating']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title('Heatmap of Correlation Between Price and Rating')
plt.tight_layout()
plt.show()
```
